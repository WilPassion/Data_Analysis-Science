```python
import pandas as pd

mkt_data = pd.read_csv('datasets/marketing_data.csv', delimiter=';')
mkt_data.head(27)
```

```python
mkt_data = mkt_data.rename(columns= {
    'ID': 'ID',
    'Ano Nascimento': 'Year of Birth',
    'Escolaridade': 'Education Level',
    'Estado Civil': 'Marital Status',
    'Salario Anual': 'Annual Salary',
    'Filhos em Casa': 'Children at Home',
    'Adolescentes em Casa': 'Teenagers at Home',
    'Data Cadastro': 'Registration Date',
    'Dias Desde Ultima Compra': 'Days Since Last Purchase',
    'Gasto com Eletronicos': 'Spending on Electronics',
    'Gasto com Brinquedos': 'Spending on Toys',
    'Gasto com Moveis': 'Spending on Furniture',
    'Gasto com Utilidades': 'Spending on Utilities',
    'Gasto com Alimentos': 'Spending on Food',
    'Gasto com Vestuario': 'Spending on Clothing',
    'Numero de Compras com Desconto': 'Number of Discount Purchases',
    'Numero de Compras na Web': 'Number of Web Purchases',
    'Numero de Compras via Catalogo': 'Number of Catalog Purchases',
    'Numero de Compras na Loja': 'Number of Store Purchases',
    'Numero Visitas WebSite Mes': 'Number of Website Visits per Month',
    'Compra na Campanha 1': 'Purchase in Campaign 1',
    'Compra na Campanha 2': 'Purchase in Campaign 2',
    'Compra na Campanha 3': 'Purchase in Campaign 3',
    'Compra na Campanha 4': 'Purchase in Campaign 4',
    'Compra na Campanha 5': 'Purchase in Campaign 5',
    'Comprou': 'Purchased',
    'Pais': 'Country'
})

print(mkt_data.head(27))
```

```python
education_level_translation = {
    'Mestrado': "Master's",
    'Doutorado': 'Doctorate',
    'Ensino Médio': 'High School',
    'Bacharelado': "Bachelor's",
    'Ensino Fundamental': 'Basic Education'
}

marital_status_translation = {
    'Solteiro': 'Single',
    'Casado': 'Married',
    'Divorciado': 'Divorced',
    'Viúvo': 'Widowed'
}

country_translation = {
    'Chile': 'Chile',
    'Espanha': 'Spain',
    'Argentina': 'Argentina',
    'Alemanha': 'Germany',
    'Estados Unidos': 'United States',
    'Brasil': 'Brazil',
    'Portugal': 'Portugal'
}

mkt_data['Escolaridade'] = mkt_data['Escolaridade'].map(education_level_translation)
mkt_data['Estado Civil'] = mkt_data['Estado Civil	'].map(marital_status_translation)
mkt_data['Pais'] = mkt_data['Pais'].map(country_translation)

print(mkt_data.head(27))
```

```python
mkt_data.to_csv('datasets_transformed/translated_marketing_data.csv', index=False)
```
